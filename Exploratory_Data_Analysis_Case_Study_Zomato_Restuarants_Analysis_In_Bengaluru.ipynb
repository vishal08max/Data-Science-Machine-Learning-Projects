**What is EDA?**

**Exploratory Data Analysis (EDA)** is used by data scientists to analyze and investigate data sets and summarize their main characteristics, often employing data visualization methods.

EDA helps determine how best to manipulate data sources to get the answers you need, making it easier for data scientists to discover patterns, spot anomalies, test a hypothesis, or check assumptions.

EDA is primarily used to see what data can reveal beyond the formal modeling or hypothesis testing task and provides a provides a better understanding of data set variables and the relationships between them. It can also help determine if the statistical techniques you are considering for data analysis are appropriate. Originally developed by American mathematician John Tukey in the 1970s, EDA techniques continue to be a widely used method in the data discovery process today.

We can see that no values are given in above menu_item column so we are removing that column which is not suitable for our further analysis.

```python
data.drop(['menu_item'],axis=1,inplace=True)
```

```python
data.columns
```

So further we have rate column which is in the object datatype, we are converting that to the float datatype for the further analysis.

```python
data.rate.unique()
```

To remove the /5 we have written a python function below which removes /5 which is not needed for our analysis.

```python
def handlerate(value):
  if(value == 'NEW' or value == '-'):
    return np.nan
  else:
    value = str(value).split('/')
    value = value[0]
    return float(value)

data['rate'] = data['rate'].apply(handlerate)
```

```python
data['rate'].unique()
```

```python
data['rate'].head()
```

```python
data.rate.isna().sum()
```

```python
data['rate'].fillna(data['rate'].mean(),inplace=True)
data['rate'].isna().sum()
```

```python
data.info()
```

Soo moving on to the next cleaning stage we have noticed that some columns such as url, address, phone, reviews_list are not needed for our analysis part, so we are going to remove these columns from our dataset.

```python
data = data.drop(['url','address','phone','reviews_list'],axis=1)
data.head()
```

```python
data.shape
```

```python
data.columns
```

Above we are containing two columns related to the locations such as location and listed_in(city). but we are going to use only one column for our analysis. So we are dropping the listed_in(city) column.

```python
data = data.drop(['listed_in(city)'],axis=1)
```

```python
data.head()
```

Sooo here we can rename the above columns approx_cost(for two people) and listed_in(type) to the new names as Cost2Persons and Type.

```python
data.rename(columns = {'approx_cost(for two people)':'Cost2Persons','listed_in(type)':'Type'}, inplace=True)
data.head()
```

```python
data.info()
```

Here in the above information of every column which is having its own correct datatype format but when we observe the column 'Cost2Persons' which should be the integer datatype but its having object soo we need to convert that to integer datatype.

```python
data.Cost2Persons.unique()
```

When we check the values in the above column, comma which is seperating the values therefore considered as a string by interpreter sooo we have to change the datatype to integer.

```python
def handlecomma(value):
  value = str(value)
  if ',' in value:
    value = value.replace(',','')
    return float(value)
  else:
    return float(value)
```

After writing the python function we will apply that function to our column.

```python
data['Cost2Persons'] = data['Cost2Persons'].apply(handlecomma)
```

As we can see that the comma is removed.

```python
data.Cost2Persons.unique()
```

```python
data.info()
```

```python
data['Cost2Persons'] = data['Cost2Persons'].astype('int')
```

We have converted the float datatype to the integer once again for our further analysis.

```python
data.head()
```

```python
data.info()
```
